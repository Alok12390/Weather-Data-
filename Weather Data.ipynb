```python

df = pd.read_csv("Weather Data.csv")


df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='mixed')


df_daily = df.set_index('Date/Time').resample('D')['Wind Speed_km/h'].mean().reset_index()
plt.figure(figsize=(14, 5))
plt.plot(df_daily['Date/Time'], df_daily['Wind Speed_km/h'], color='royalblue', linewidth=1.5)
plt.title('Daily Average Wind Speed Trend Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Wind Speed (km/h)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('weather_line.png')

plt.show()
```

```python
# ===================================================================
# GRAPH 3: Wind Speed Across Top Weather Conditions (Box Plot)
# ===================================================================
# Filter for the top 5 most common weather conditions to avoid a messy chart
top_weather = df['Weather'].value_counts().nlargest(5).index
df_filtered = df[df['Weather'].isin(top_weather)]

plt.figure(figsize=(12, 6))

# Updated line with hue added:
sns.boxplot(data=df_filtered, x='Weather', y='Wind Speed_km/h', palette='Set2', hue='Weather')

plt.title('Wind Speed Spread Across Top 5 Weather Conditions', fontsize=14, fontweight='bold')
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Wind Speed (km/h)', fontsize=12)
plt.xticks(rotation=15)
plt.tight_layout()
plt.savefig('t5weathercondition_box.png')

plt.show()
```
